```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,
         color='red',
         linestyle='-',
         marker='o',
         linewidth=2,
         markersize=10)

plt.grid(True)
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9, 5))
plt.subplot(1,2,1)
plt.plot(x, y1, color='green')
plt.title('Plot 1')
plt.show()

plt.subplot(1,2,2)
plt.plot(x, y2, color='green')
plt.title('Plot 2')
plt.show()
```
