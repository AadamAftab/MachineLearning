```python
import numpy as np
```

```python
#1D array using numpy
arr1=np.array([1,2,3,4,5])
print(arr1)
print(type(arr1))
print(arr1.shape) #dimension of array
```

```python
arr2=np.array([1,2,3,4,5])
arr2.reshape(1, 5) # 1 row and 5 columns 
```

```python
arr3=np.array([[1,2,3,4,5], [2,3,4,5,6]])
print(arr3.shape)
```

```python
np.arange(0, 10, 2).reshape(1, 5)
```

```python
np.ones((3,4))
```

```python
#identity matrix
np.eye(3,3)
```

```python
arr4=np.array([[1,2,3], [4,5,6]])
print(arr4.shape) #shape of array
print(arr4.ndim) #row dimn
print(arr4.size) #no of elements
print(arr4.dtype) #element data type
print(arr4.itemsize) #size of element
```

```python
arr1=np.array([1,2,3,4,5])
arr2=np.array([-1,-2,-3,-4,-5])
print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(arr1/arr2)
```

```python
#universal functions
arr=np.array([2,3,4,5,6])
print(np.sqrt(arr))
print(np.exp(arr))
print(np.sin(arr))
print(np.log(arr))
```
