```python
def multiples(n):
    lst=[]
    for i in range(1, 6):
        lst.append(n*i)
    return lst
```

```python
d4={}
for i in range(1, 6):
    d4[i] = multiples(i)
print(d4)
```

```python
# Create dictionary with first 5 positive integers and their squares
d = {}

for i in range(1, 6):
    d[i] = i ** 2

# Swap keys and values
new_d = {}

for key, value in d.items():
    new_d[value] = key

# Print new dictionary
print(new_d)
```

```python
from collections import defaultdict

# Create a default dictionary with empty list as default value
d = defaultdict(list)

# Add elements to the lists
d["fruits"].append("apple")
d["fruits"].append("banana")

d["numbers"].append(10)
d["numbers"].append(20)

# Print the dictionary
print(dict(d))
```
