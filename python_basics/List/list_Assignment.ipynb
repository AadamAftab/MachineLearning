```python
#[1,2,3,4] --> [4,1,2,3] 
l17=[1,2,3,4]
#[1,2,3,4] -> [3,2,1,4] -> [4,1,2,3]
n=int(input())%len(l17)
l17[0:len(l17)-n] = l17[0:len(l17)-n][::-1]
l17[len(l17)-n:] = l17[len(l17)-n:][::-1]
l17=l17[::-1]
print(l17)
```
